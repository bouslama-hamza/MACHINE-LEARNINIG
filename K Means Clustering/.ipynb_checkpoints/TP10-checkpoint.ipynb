#### ***Q1***-utilisez la fonction ***make_blobs(n_samples,centers, cluster_std,random_state)*** du module ***sklearn.datasets*** pour generer trois classes puis visualiser ces classes 



```python
from sklearn.datasets import make_blobs
X, Y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0] , X[:,1] , s =50 , c = Y)
```

#### ***Q2***  Ecrire un script pour pr√©dire la classe de x avec l'algorithme du k-means puis visualiser ces trois classe avec les cluster_centers

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0] , X[:,1] , s =50 , c = Y)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
wcss = [] 
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, random_state = 42)
    kmeans.fit(X) 
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

#### Q3. A quelle classe apartient les trois points suivants: x1=(-2,1); x2(0,1) ;x3=(2.5,5)  

```python
import numpy as np
x_new = np.array([[-2,1] , [0,1] , [2.5 , 5]])
y_pred = kmeans.predict(x_new)
y_pred
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
X = data.data
```

```python
from sklearn.cluster import KMeans
```

```python
import pandas as pd
x = pd.DataFrame(data.data)
x.columns = ['Sepal_Length','Sepal_width','Petal_Length','Petal_width']
y = pd.DataFrame(data.target)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(x)
```

```python
plt.scatter(x.Petal_Length, x.Petal_width ,c =y)
```

```python
wcss = [] 
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, random_state = 42)
    kmeans.fit(x) 
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```
